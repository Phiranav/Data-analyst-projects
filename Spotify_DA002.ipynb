```python
# importing essential libraries

import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
# loading data from local system

df_tracks = pd.read_csv("X:\\python project\\datasets\\archive\\tracks.csv")
df_tracks.head(10)
```

```python
#checking for null values & details of dataset:
df_tracks.isnull().sum()
df_tracks.info()
```

```python
#least song
sort_df = df_tracks.sort_values('popularity', ascending = True).head(10)
sort_df
```

```python
df_tracks.describe().T
```

```python
famous_song = df_tracks.query('popularity > 95')\
.sort_values('popularity', ascending = False)

famous_song.head(10)


```

```python
# fromatting date and setting new indexes


#df_tracks.set_index("release_date", inplace = True)
df_tracks.index = pd.to_datetime(df_tracks.index, format = 'mixed')
df_tracks.head()
```

```python
df_tracks[["artists"]].iloc[[2]]
```

```python
#converting millisecond to sec

df_tracks["duration"] = df_tracks["duration_ms"]\
.apply(lambda msec: round(msec/1000))

df_tracks.drop("duration_ms", inplace = True, axis = 1)



```

```python
df_tracks.duration.head()
```

```python
#removing all string valued columns for correlation calculation

corr_df = df_tracks.drop(['key','id','name','artists','id_artists'], axis = 1)\
.corr(method ='pearson')
```

```python
#Visulation for Correlation using heatmap

plt.figure(figsize= (20,8))
heatmap =sns.heatmap(corr_df, annot = True, fmt = '.3g', vmin = -1, vmax = 1, center =0, cmap ="crest",linewidth =1)
heatmap.set_title("Correlation Heatmap Between Variable")
plt.show()
```

```python
#calculating correlation between loudness and energy

sample_df = df_tracks.sample(int(0.004*len(df_tracks)))
print(len(sample_df))
```

```python
plt.figure(figsize=(15,7))
sns.regplot(data = sample_df, x = "energy", y = "loudness", color = 'c', line_kws = {'color':'b'})\
.set(title = "loudness Vs Energy")
```

```python
# popularity vs acousticness

plt.figure(figsize=(15,7))
sns.regplot(data = sample_df, x = "popularity", y = "acousticness", color = 'g', line_kws = {'color':'r'})\
.set(title = "Popularity Vs Acousticness")
plt.show()
```

```python
df_tracks['dates'] = df_tracks.index.get_level_values('release_date')
df_tracks.dates = pd.to_datetime(df_tracks.dates)
years = df_tracks.dates.dt.year
df_tracks.columns
```

```python
df_tracks['dates'].tail()
```

```python
#distribution plots
sns.displot(data = years, discrete = True, aspect = 2, height = 6, kind = 'hist')\
.set(title = "No.of.songs per year")
plt.show()
```

```python
#Duration by years

total_dr = df_tracks.duration
fig_dims = (20,8)
fig, ax = plt.subplots(figsize = fig_dims)
fig = sns.barplot(x = years, y = total_dr, color = 'orange').set(title = "year vs duration")
plt.xticks(rotation = 90)
plt.show()




```

```python

total_dr = df_tracks.duration
sns.set_style(style= "whitegrid")
fig_dims = (20,7)
fig, ax = plt.subplots(figsize = fig_dims)
fig = sns.lineplot(x = years, y = total_dr, color = '#a218ed').set(title = "year vs duration")
plt.xticks(rotation = 90)
plt.show()
```

```python
#converting secs into mintues to find podcasts and songs
df_tracks['Duration_min'] = df_tracks['duration']/60.
```

```python
df_tracks['category'] = df_tracks['Duration_min'].round().apply( lambda min : 'podcast' if min > 4 else 'song')
```

```python
df_tracks['category']
```

```python
count_songs = df_tracks['category'].value_counts()

plt.figure(figsize = (8,8))

plt.pie(count_songs, labels = count_songs.index, startangle= 90, autopct = '%1.1f%%' )
plt.show()
```

```python
# creating songs vs podcast

plt.figure(figsize = (6,1))
category_popularity = df_tracks.groupby('category')['popularity'].sum()

bar_color = ['#e67620','#63615f']

plt.barh(category_popularity.index , category_popularity.values, color = bar_color )
plt.xlabel('Popularity')
plt.ylabel('category')
plt.legend()
plt.show()
```
