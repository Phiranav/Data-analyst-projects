```python
Coaster = df2['year_introduced'].value_counts().head(10) \
.plot(kind='bar', title = 'Top Years coasters Introduced')
Coaster.set_xlabel('Year introduced')
Coaster.set_ylabel('count')
```

```python
#what are the location with the fastest roller coaster (minimum of 10 coasters)?

coaster_details = df2.query('Location !="Other" ').groupby('Location')['speed_mph']\
.agg(['mean','count']).query('count>=10')\
.sort_values('mean')['mean']\
.plot(kind = 'barh',figsize=(15,6), title = "Average speed of Roller Coaster by Location")

coaster_details.set_xlabel('Average Coaster Speed')
plt.show()
```
